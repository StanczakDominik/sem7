## Równania temperatury i entropii
* z równania energii wewnętrznej
* przez związki termodynamiczne (Maxwella)

Z termodynamiki i doświadczenia: do opisu stanu dowolnej masy płynu w równowadze lokalnej (przy stałej liczbie cząstek w tej masie, więc układ kanoniczny), przestrzennie jednorodnej i niezmienniczej w czasie **wystarczą dwa parametry stanu**, na przykład (wybór dość dowolny)
* gęstość i ciśnienie
* ciśnienie i objętość właściwa ($V=1/\rho$)

Jednym z najważniejszych parametrów stanu jest temperatura. Jej związek z $p, V$ (lub $p, \rho$) nazywamy **równaniem stanu**. W przestrzeni 3D ($p, T, V$) określa takowe powierzchnię. 

### 1 Zasada Termodynamiki
1. Istnieje **funkcja stanu** zwana energią wewnętrzną $U$ (zmiana $U$ między stanami **nie zależy od drogi** między stanami)
2. Zmiana energii wewnętrznej następuje przez pracę mechaniczną i wymianę ciepła z otoczeniem.
$$T dS - p dV = dU$$
Nie używamy $\mu dN$ bo mamy stałą liczbę cząstek, $dN = 0$.
"Makroskopowo" (znaki dodatnie gdy dostarczamy energię do układu, ciepło dostarczone Q i praca wykonana W nie są funkcjami stanu):
$$\Delta E = Q + W$$

Można zapisać $dE = dQ + dW$ z kreślonymi $d$ w $dQ$, $dW$ - są to formy Pfaffa (nie różniczki zupełne).

### 2 zasada termodynamiki
1. Istnieje funkcja stanu zwana entropią $S$
2. W infinitesymalnym procesie odwracalnym $dQ$ jest proporcjonalne do różniczki funkcji stanu:
$$dQ = TdS$$

$$dS = dQ / T$$

więc $T^{-1}$ jest czynnikiem całkującym dla dQ (po podzieleniu przez T dostajemy różniczkę zupełną).

### Związki Maxwella
$E, T, S$ etc to funkcje stanu. Dlatego ich różniczki to różniczki zupełne:
$$dE = TdS - pdV = (\frac{\partial E}{\partial S})_V dS + (\frac{\partial E}{\partial V})_S dV$$

Teraz liczymy pochodne mieszane i z tw. Schwarza je przyrównujemy:
$$(\frac{\partial T}{\partial V})_S = -(\frac{\partial p}{\partial S})_V$$

### Inne potencjały termodynamiczne:
* entalpia $ H = E + pV, dH = TdS + Vdp$
* energia swobodna $F = E - TS, dF = -S dT -p dV$
* entalpia swobodna Gibbsa, $G = E pV - TS, dG = -SdT + V dp$

Pozostałe tożsamości Maxwella liczymy analogicznie:

* $T_p (S const) = V_s (p const)$
* $S_V (T const) = p_T (V)$
* $-S_p (T const) = V_T (p const)$

wykorzystując trzy pozostałe potencjały.
### Ciepło właściwe na jednostkę masy
$$C_\alpha = (\frac{dQ}{dT})_\alpha = T (\frac{\partial S}{\partial T})_\alpha$$
gdzie pochodną bierzemy na kierunku danego procesu.

Czyli

$$C_p = (\frac{dQ}{dT})_p = T (\frac{\partial S}{\partial T})_p$$
$$C_V = (\frac{dQ}{dT})_V = T (\frac{\partial S}{\partial T})_V$$

$$dS = dS(T, p) = \frac{C_p}{T} dT - (\frac{\partial V}{\partial T})_p dp$$
Współczynnik rozszerzalności cieplnej.
$$\beta = \frac{1}{V} (\frac{\partial V}{\partial T})_p$$

$$T dS = dE + pdV = c_p dT - \beta V T dp$$

wzór wyrażający $dE$ przez wielkości bezpośrednio mierzalne.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 0.5
r = np.linspace(0.001, 1, 100)
u = np.zeros_like(r)
Omega = 1
inside = r<a
outside = ~inside
u[inside] = Omega*r[inside]
u[outside] = Omega*a**2/r[outside]
plt.plot(r, u)
```

```python
theta = np.linspace(0, 2*np.pi, 200)
r_grid, theta_grid = np.meshgrid(r, theta)
u_r_grid = np.zeros_like(r_grid)
u_theta_grid = np.zeros_like(r_grid)
u_theta_grid[:] = u
```

```python
x_grid = r_grid * np.cos(theta_grid)
y_grid = r_grid * np.sin(theta_grid)
u_x_grid = u_theta_grid * - np.sin(theta_grid)
u_y_grid = u_theta_grid * np.cos(theta_grid)
```

```python
plt.quiver(x_grid, y_grid, u_x_grid, u_y_grid)
```
