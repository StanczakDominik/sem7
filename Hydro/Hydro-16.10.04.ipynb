### Mechanika Ośrodków Ciągłych, od najogólniejszego opisu:
1. Mechanika statystyczna: opis cząstek poprzez podanie położeń i pędów wszystkich cząstek układu. Ogromna liczba stopni swobody
    * sypie się w złożonych układach
2. Teoria kinetyczna: funkcja rozkładu $f(x, p, t): 3+3+1=7$ stopni swobody. $f(x, p, t) dp dx$ - liczba cząstek w obszarze przestrzeni fazowej $dp dx$.
3. M.O.C. - zakładamy, że w każdym punkcie przestrzeni rozkład w przestrzeni prędkości jest równowagowy *(Maxwelliański?)*. Każdemu otoczeniu danego punktu zakładamy że możemy sensownie przypisać $T, p, \rho$ i inne wielkości termodynamiczne.
    * Słowem: $\rho(x), t), T(x, t), u(x, t)$ ($u$: prędkość płynu). 
 
Notacja wykładowa: podkreślenie oznacza wektor, kreska u góry: średnią. Jeśli jest to jednoznaczne, tutaj będę osobiście zostawiał wektory bez oznaczenia.

* Fale uderzeniowe: nieciągłości gęstości (prędkość fali $\sim \rho$). Poza falą można **zszyć** dwa rozwiązania.

Mamy też zestaw parametrów empirycznych, jak np. lepkość, przewodnictwo cieplne. W prostych przypadkach można wyprowadzić, w rzeczywistych - doświadczalnie.

MOC dzieli się na:
* Elastomechanikę (teorię sprężystości, plastyczności ciał stałych)
* Hydrodynamikę (teorię przepływu cieczy)
* Aerodynamikę (teorię przepływu gazów)

Dwa ostatnie łączą się w mechanikę płynów. Różnica: wartość ściśliwości, dla gazów jest spora, dla cieczy prawie żadna.

### Układy nieco bardziej złożone:
* MHD (magnetohydrodynamika) - przepływy płynów przewodzących prąd elektryczny, sprzężone z polem magnetycznym (plazemki na ten przykład). Siła Lorentza - podstawowa, poza tym równania Maxwella. Prawo Ohma domyka układ równań. Kluczowe dla geoastrofizyki - "astronomiczne" (planetarne, gwiezdne) pole magnetyczne generalnie pochodzi zawsze od jakiejś warstwy płynu, który się rusza.

Warstwy ziemi od zewnątrz:
* 3000km grubości: płaszcz
* 2200km: płynne żelazo
* 1000km: super gęste żelazo
Jest to pewna wersja prądnicy: odpowiednik kręcenia to ciepło pierwotne (dosłownie pierwotne), strumień ciepła skierowany na zewnątrz napędza ruchy konwekcyjne ($\Delta T \sim 1100 K$). To powoduje wykrzywianie pola magnetycznego -> prąd -> konwekcję -> sprzężenie zwrotne -> wielkoskalowe pole magnetyczne

* Przepływy wielofazowe *(ooo, multiphase!)* - więcej niż jedna faza - stała, ciekła, gazowa. Np. zawiesiny, niemieszające się płyny, aerozole w atmosferze.

* Ciało stałe - mała deformacja w liniowym przybliżeniu spowoduje naprężenie, zaś po odjęciu siły układ wróci do pierwotnego stanu (jeśli jest sprężysty, oczywiście). Interesuje nas deformacja tego ciała.

* Ciecz: interesuje nas przepływ. Dowolnie mała siła spowoduje duże przemieszczenie.

Ciecze nie są ściśliwe (najwyżej zaniedbywalnie). Gazy jak najbardziej są.

* Ciało stałe: uporządkowanie długozasięgowe, trwałe
* Ciecze: uporządkowanie średniozasięgowe, nietrwałe (*chwilowe* domeny w cieczach)
* Gazy: brak uporządkowania

Aby dobrze zdefiniować gęstość $\rho$, musimy wziąć odpowiednio duży obszar, by $\rho$ nie fluktuowało zbyt znacząco *(jeśli wzięcie jeszcze jednej cząstki do obszaru powoduje znaczący skok $\rho$, to chyba bierzemy za mały obszar)*.

Wybór odpowiedniego obszaru to tzw. **hipoteza continuum**, tycząca się w ten sam sposób $T(x, t), p(x, t), u(x, t)$. *Innymi słowy: zakładamy, że elementy płynu tworzą ciągły ośrodek, pomijamy efekty od poszczególnych cząstek, mimo tego, że faktycznie każdy płyn składa się właśnie z cząstek.*

Chwilowa gęstość: mamy sześcienny obszar $\Delta V$, gęstość to $\frac{\Delta m(t)}{\Delta V}$, $r = (\Delta V)^(1/3)

Gęstość definiujemy na *plateau* - w obszarach fluktuacji nie zachodzi równowaga, dostajemy bezsensowne wyniki. Ten koncept to **separacja skal przestrzennych**.

**Hipoteza continuum** zakłada właśnie istnienie plateau.

### M.O.C. nadaje się do opisu zjawisk:
1. *wolnych, czyli* o skali czasowej dużej w porównaniu z czasem relaksacji ($\tau \sim 10^{-11} s$ w normalnych warunkach)
2. Skali przestrzennej dużej w porównaniu ze średnią drogą swobodną ($\lambda \sim 10^{-9} m$ w warunkach normalnych, $\lambda \sim 10^{13}$ w gazie międzywiezdnym - choć tam często przydaje się teoria kinetyczna).

### Równanie ciągłości, AKA prawo zachowania masy

$\rho(x, t), u(x,t)$ - gęstość i prędkość w chwili t. $V$ - obszar ustalony w układzie odniesienia laboratorium.

**F2.png**

Masa zmienia się w czasie przepływu, jeśli gęstość nie jest jednorodna

$$m\text{ wewnątrz V} = \int \int \int_V \rho d^3r$$

Masę wchodzącą do obszaru V w czasie $\delta t$ przez mały element powierzchni $\delta A$ oznaczamy $\delta m$.

$$\delta m  = - n \cdot u \rho \delta A \delta t$$

($-$ stąd, że wektor normalny $n$ jest skierowany na zewnątrz, a masa wchodzi do środka).

Zatem:

$$\frac{d}{dt} \int_V \rho d^3r = -\int_{\partial V} \rho u \cdot n dA + \text{pominięte źródła masy}$$

Z twierdzenia Gaussa (lub Stokesa 3D)

$$\frac{d}{dt} \int_V \rho d^3r = - \int_V \nabla \cdot (\rho u) d^3r$$

Wchodzimy z pochodną po całkę, stosując pochodną Eulerowską (możemy tak zrobić, bo V jest określone jako konkretny obszar w laboratorium, więc teoretycznie stacjonarny

$$\int_V \frac{\partial \rho}{\partial t} d^3r = - \int_V \nabla \cdot (\rho u) d^3r$$

Czyli dostajemy **całkową postać prawa zachowania masy**:

$$\int_V \Big( \frac{\partial \rho}{\partial t} + \nabla \cdot (\rho u) \Big) d^3r = 0 $$

Jest to spełnione dla każdego V, a co za tym idzie dostajemy **różniczkową postać równania ciągłości**:

$$\frac{\partial \rho}{\partial } + \nabla \cdot (\rho u) = 0 + \text{ewentualna gęstość źródeł}$$

$$\frac{\partial \rho}{\partial t} = - \nabla \cdot (\rho u)$$

Forma tego tego prawa zachowania jest uniwersalna (np. bilans ciepła), np:

$$\frac{\partial T}{\partial t} = - \nabla \cdot \vec{Q} + S$$
* $S$ - źródła ciepła, np. radioaktywne.
* $$Q = T\vec{u} - \kappa \nabla T$$
* $uT$ - przepływ ciepła - człon adwekcyjny (unoszenia).
* $-\kappa \nabla T$ - przepływ ciepła - człon molekularny - to co czujemy przez łyżkę w herbacie, garnek na gazie...

$$\frac{\partial T}{\partial t} + \nabla \cdot (u T) = \kappa \nabla^2 T + S$$

Przy zerowej prędkości i braku źródełek radioaktywnych dostaniemy samo Brownowskie równanie dyfuzji, w tym przypadku: molekularnej ciepła.

### Opisy Eulera i Lagrange'a
W opisie Eulera występuje przestrzeń fizyczna $(x, t)$ zaś w Lagrange'a przestrzeń ciała $(a, t)$

$$\frac{dx_1}{dt} = u_1(x,t)$$
$$\frac{dx_2}{dt} = u_2(x,t)$$
$$\frac{dx_3}{dt} = u_3(x,t)$$
Układ równań różniczkowych zwyczajnych opisujących trajektorię cząstki, *Pathlines* lub *Trajectories*

Z warunku początkowego, dla każdej oddzielnej cząstki:
$$x(t=t_0)=a$$

Rozwiązanie: $x(t, t_0, a)$ - krzywa opisująca trajektorię. Dzięki niej można zrobić zamianę zmiennych $(a, t_0) \iff (x, t)$ *(trajektorie nie kolapsują się jedna na drugą)*

Dla każdego $t$ mamy zamianę zmiennych w $R^3$.

$\frac{\partial}{\partial t}$ - pochodna po czasie przy ustalonym $x$.

$\frac{D}{Dt} = \frac{d}{dt}$ - pochodna **substancjalna** przy ustalonym $a$ (wskazaniu czy *indexie* cząstki).

### Dygresja na wizualizację: cięcia Poincare

Chcemy narysować w przestrzeni rozwiązania układu równań na trajektorię. Wybieramy sobie płaszczyznę i tam, gdzie trajektoria ją przecina, stawiamy kropkę. Zbiór kropek to cięcie Poincare. 

Rozwiązaniem równań trajektorii jest odwzorowanie: $t->x(t, t_0, a)$ - trajektoria cząstki płynu, która w chwili $t=t_0$ była w punkcie $a$

### Równania linii całkowych **chwilowego** pola prędkości (*streamlines*)
Zbiór stycznych do wektorów pola prędkości

**F3.PNG**

$$\frac{d x_1}{d\xi} = u_1(x, t)$$
$$\frac{d x_2}{d\xi} = u_2(x, t)$$
$$\frac{d x_3}{d\xi} = u_3(x, t)$$

**Różnica między trajektorią a linią prądu jest taka, że linia prądu jest w danej chwili, jej zmienną niezależną jest $\xi$, a parametrem czas. W trajektorii - przeciwnie, zmienną niezależną jest właśnie czas.**

$\xi -> x(\xi, t, x_0)$ - linia prądu w chwili $t$ przechodząca przez $x_0$ - **nie przez $a$, nie jest związana z konkretną cząstką, a z punktem w przestrzeni!**

Koncepcyjnie: w danej chwili robimy zdjęcie polu prędkości. Rozwiązanie to krzywa całkowa pola prędkości, czyli linie do ktorych styczne są wektory pola w danej chwili.


Przy linii prądu mamy zależność tylko od $\xi-\xi_0$, dowolnie wybranego "poziomu odniesienia", ponieważ prawe stron równań nie zależą od $\xi$. W przypadku trajektorii mamy zależność indywidualnie od $t$, $t_0$.

Trzeci typ rozpatrywanych krzwych bierze się z trajektorii i nazywa się **smugami**.

Smuga to odwzorowanie:
$ t_0 -> x(t_0, t, a)$ - tym razem na rozwiązanie równań trajektorii patrzymy jako na funkcję t_0 - $t$ jest ustalone, robimy snapshot

**F4.PNG**

Nasza krzywa to $x(t_0, t, a)$ - cząstki w różnych chwilach t_0 były w punkcie $a$.

W ogólności jeżeli $u$ zależy od czasu te 3 linie (trajektorie, linie prądu, smugi) mogą być zupełnie różne.

Jeżeli u nie zależy od czasu, układy równań na trajektorie i na linie prądu są tożsame. Te trzy linie wyglądają wówczas tak samo **przy przepływie stacjonarym.**

$$\frac{dx}{dt} = u(x, (t)), \frac{dx}{d\xi} = u(x, (\xi))$$
