Inne napędzanie przepływu - nie przez warunek brzegowy (ruch ścianki, a przez gradient ciśnienia).

Poszukujemy rozwiązania osiowo symetrycznego:

$$U = u(r) \hat{e}_z$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
z = np.linspace(-4, -0.001, 10000)
# k = np.sqrt(Omega / ni)
k = 1/0.39
S = 1

main = S/(k*np.sqrt(2))*np.exp(k*z)
u = main * np.cos(k*z - np.pi/4)
v = main * np.sin(k*z - np.pi/4)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(u, v, z)
```
