# jak wygląda tensor naprężeń w cieczy:

W cieczy spoczywającej (nieruchomej)

$$\sigma_{ij} = -p \delta_{ij}$$
$$p = - Tr(\sigma)/3$$

Ciśnienie ujemne: rozprężanie. Ciśnienie dodatnie - ściskanie.

W poruszającej się cieczy wyróżniamy część izotropową i anizotropową (dysypatywną, którą dorzucamy do poprzedniego) $d_{ij}$. p - średnie naprężenie normalne (po wszystkich kierunkach; w spoczynku ciecz nie zmienia kszałtu, naprężenie na każdym kierunku jest takie samo).

### Anizotropowa część tensora naprężeń

$d_{ij}$ - anizotropowa (dysypatywna) część $\sigma_{ij}$ związana z występowaniem naprężeń ścinających. $Tr(d_{ij}) = 0$, jeśli znika tzw. druga lepkość (objętościowa).

$d_{ij}$ - zagadnienie **transportu**, transport molekularny między warstwami płynu *jak skoki między liniami pola magnetycznego*.

$d_{ij}$ jest siłą na powierzchnię i wiąże się z transportem pędu przez tę powierzchnię.

Założenie analogiczne do prawa Ficka, ew. Fouriera:

$$\vec{Q} = -\kappa \nabla T$$

Transport ciepła jest proporcjonalny do bodźca, czyli gradientu T.

Najbardziej ogólna postać związku liniowego:

$$Q_i = C_{ij} \frac{\partial T}{\partial x_j}$$

W prawie Ficka $C_{ij} = - \kappa \delta_{ij}$.

Analogiczne założenie robimy w hydrodynamice. Przyczyną $d_{ij}$ jest gradient pola prędkości, tensor Grad $\vec{u}$. (jak prędkość jest stała w przestrzeni to robimy transformację Galileusza do układu ruszającego się z $\vec{u}$ i mamy ciecz w spoczynku).

$$d_{ij} = A_{ijkl} \frac{\partial u_k}{\partial x_l}$$

Najogólniejsza postać liniowego związku. Tensor czwartego rzędu $A$ nie zależy od gradientu $\vec{u}$, ale może zależeć od stanu cieczy.

Część symetryczna Grad $\vec{u}$ ma 6 niezależnych składowych: $e_{kl} = 0.5 (\frac{\partial u_k}{\partial x_l} + \frac{\partial u_l}{\partial x_k})$ - tensor prędkośći deformacji, symetryczny

 Część antysymetryczna $e_{kl} = 0.5 (\frac{\partial u_k}{\partial x_l} - \frac{\partial u_l}{\partial x_k})$, zera na diagonali, 3 niezależne składowe, i to można wyrazić rotacją jako $ = -0.5 \epsilon_{ikl} \omega_l$, gdzie $\vec{\omega} = \nabla \times \vec{u}$.
 
 Mamy zatem: $\frac{\partial u_k}{\partial x_l} = e_kl - 0.5 \epsilon_{ikl} \omega_l$
 
 $d_{ij} = A_{ijkl} e_{kl} - 0.5 A_{ijkl} \epsilon_{klm} \omega_m$
 
 Jeśli ciecz jest izotropowa, czyli jej własności fizyczne nie zależą od kierunku, to $A_{ijkl} też jest izotropowe - jest to założenie.
 
 Izotropowy (numeryczny wg teorii grup) tensor drugiego rzędu ma postać: $const * \delta_{ij}$. Izotropowy tensor czwartego rzędu ma postać $A_{ijkl} = \mu * \delta_{ik}\delta_{jl} + \mu' \delta_{il}\delta_{jk} + \mu'' \delta_{ij}\delta_{lk}$. To jest najogólniejsza forma! Współczynniki $\mu$ mogą zależeć od stanu cieczy, np. temperatury.
 
Ponieważ $\sigma_{ij} = \sigma_{ji}$, to $A_{ijkl} = A_{jikl} \to \mu = \mu'$.

Czyli: $A_{ijkl} \epsilon_{klm} = 0$, bo $A$ jest symetrycnze w $kl$,zaś $\epsilon_{klm}$ antysymetryczny w $kl$. Wobec tego $d_{ij} = A_{ijkl} e_{kl}$. Podstawiając

$$d_{ij} = 2 \mu e_{ij} + \mu'' \nabla \cdot \vec{u} \delta_{ij}$$

Nawiasem mówiąc $\nabla \cdot \vec{u} = Tr(e) = e_kk$, zaś $Tr(Grad \vec{u}) = Tr(Grad\vec{u})$ = \nabla \cdot \vec{u}$.
